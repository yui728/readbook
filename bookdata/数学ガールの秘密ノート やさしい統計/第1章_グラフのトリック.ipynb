## 1.3 大きく見せたい

- 大きく見せたい時は、グラフの下を切り取る

```python
import numpy as np

fig = plt.figure()
ax1 = plt.axes([0, 0.12, 1, 0.7])
ax2 = plt.axes([0, 0, 1, 0.1], sharex = ax1)
ax1.set_title('下を切り取った折れ線グラフ')
ax1.set_ylabel('社員数(人)')
ax2.set_xlabel('年')
ax2.set_ylim(0, 1)
ax2.set_yticks([0])
ax2.spines['top'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.set_ylim(90, 150)
ax1.set_yticks([100, 130])

ax1.xaxis.tick_top()
ax2.xaxis.tick_bottom()
ax1.tick_params(labeltop = 'false')
ax2.set_aspect('auto', adjustable = 'datalim')

d = .015
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs) 

d = .015
kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (0.8 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (0.8 - d, 1 + d), **kwargs)

# 年数
x = [0, 1, 2, 3, 4, 5]
# 社員数
y = [100, 117, 126, 133, 135, 136]

# 折れ線グラフを描画
ax1.plot(x, y, marker = 'o')

plt.subplots_adjust(hspace = 0.01)

plt.show()
```

## 1.7 横軸を変える

*横軸の目盛りをいじれば、社員が増えている印象にできる*

```python
fig = plt.figure(figsize = (3, 5))
ax1 = plt.axes([0, 0.105, 1, 0.7])
ax2 = plt.axes([0, 0, 1, 0.08], sharex = ax1)
ax1.set_title('社長「社員はすごく増えてるぞ」')
ax1.set_ylabel('社員数(人)')
ax2.set_xlabel('年')
ax2.set_ylim(0, 1)
ax2.set_yticks([0])
ax2.spines['top'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.set_ylim(95, 140)
ax1.set_yticks([100, 110, 120, 130])

ax1.xaxis.tick_top()
ax2.xaxis.tick_bottom()
ax1.tick_params(labeltop = 'false')
ax2.set_aspect('auto', adjustable = 'datalim')

d = .015
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs) 

d = .015
kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (0.8 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (0.8 - d, 1 + d), **kwargs)

# 年数
x = [0, 1, 2, 3, 4, 5]
# 社員数
y = [100, 117, 126, 133, 135, 136]

# 折れ線グラフを描画
ax1.plot(x, y, marker = 'o')

plt.subplots_adjust(hspace = 0.01)

plt.show()
```

*横軸の目盛りをいじっても、データの印象が変わる*

```python
fig, ax = plt.subplots()

# 年数
x = [d for d in range(6)]
# 社員数
y = [100, 117, 126, 133, 135, 136]

ax.set_title("専務「社員はほとんど増えてません」")
ax.set_xlabel("年")
ax.set_ylabel("増加人数（人）")
ax.set_xticks(x)
ax.set_ylim(0, 150)
ax.set_yticks([d * 50 for d in range(3)])

ax.plot(x[3:], y[3:], marker = 'o')

plt.show()
```

**グラフをどう描く**かだけではなく、**データのどこを選んでグラフを描くか**も大事

```python
fig, ax = plt.subplots()

x = [d + 1 for d in range(28)]
y = [183.5, 183.9, 184,   184.2, 183.8, 184,  184.3,
     184.2, 184.1, 184.2, 184.1, 184.4, 184.4, 184.4,
     184.5, 184.6, 184.6, 184.4, 184.8, 185,   185.1,
     185.3, 185.3, 185.7, 185.5, 185.7, 186, 186.3]

ax.set_title('グラフ1「株価は継続的に上がっている？」')
ax.set_xlabel('日（2014年12月）')
ax.set_ylabel('円')
ax.set_xticks([d * 7 + 1 for d in range(5)])
ax.set_yticks([d + 183 for d in range(4)])
ax.set_ylim(182, 187)
ax.set_xlim(1, 29)

ax.plot(x, y)

plt.show()
```

## 1.9 実は、下がっていた

グラフを読む時は**《軸と目盛りをチェック》**
- グラフ1とグラフ2では、**日付の範囲が違う**
  - グラフ1：一か月分
  - グラフ2：半年分


$\therefore$グラフ1では継続的に上がっているように見えるが、グラフ2では継続的に上がっているようには見えない

## 1.10 実は、継続的に上がっていた

- グラフ1：ひと月の範囲では株価は上がっている

## 1.12 社員数のグラフを比較

```python
fig, ax = plt.subplots()

years = [d for d in range(6)]

ax.set_title('グラフ4「A社は社員が増えていない？」')
ax.set_xlabel('年')
ax.set_ylabel('社員数（人）')
ax.set_ylim(0, 3000)
ax.set_xlim(0, 5)
ax.set_yticks([d * 1000 for d in range(3)])
ax.set_xticks(years)

x = years
y_a = [100, 117, 126, 133, 135, 136]
y_b = [2210, 1903, 2089, 2020, 2052, 1950]

ax.plot(x, y_a, 'go-', label = 'A社')
ax.plot(x, y_b, 'ro-', label = 'B社')
ax.legend()

plt.show()
```

グラフを見ると、A社は横ばいのように見える
 - A社の社員数が、B社の社員数と比べて大幅に違うため
 - 補足説明としては『A社の社員数は少しずつ増えている。しかし、B社の社員数の規模と比べたらその増加人数は小さなものだ』となる

## 1.13 デッドヒートの演出

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

years = [d for d in range(6)]
x = years
y_a = [100, 117, 126, 133, 135, 136]
y_b = [2210, 1903, 2089, 2020, 2052, 1950]

ax1.set_title('グラフ5「A社とB社のデッドヒート？」')
ax1.set_ylabel('A社（人）')
ax1.set_xlabel('年')
ax2.set_ylabel('B社（人）')
ax1.set_ylim(0, 150)
ax1.set_yticks([d * 50 for d in range(3)])
ax2.set_ylim(0, 3000)
ax2.set_yticks([d * 1000 for d in range(3)])

ax1.plot(x, y_a, 'go-')
ax2.plot(x, y_b, 'ro-')
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2)

plt.show()
```

- 左側の目盛りにA社、右側の目盛りにB社でそれぞれ異なる目盛りを配置する
- A社とB社の《**社員数を比較**するグラフ》としてはアウト
- A社とB社の《社員数の**変化を比較**するグラフ》としては便利

*折れ線グラフは変化を見るのに向いている*

### 問題1-2

```python
fig, ax = plt.subplots()

x = [d + 4 for d in range(3)]
y_syokudou = [125, 90, 80]
y_restrant = [30, 40, 70]

ax.set_ylim(0, 150)
ax.set_yticks([d * 50 for d in range(3)])
ax.set_xticks(x)
ax.set_ylabel('来客数（人）')
ax.set_xlabel('（月）')

ax.plot(x, y_syokudou, 'ko-', label = '食堂A')
ax.plot(x, y_restrant, 'ko--', label = 'レストランB')
ax.legend()

plt.show()
```

1. この折れ線グラフからは来客数しか分からないため、「食堂AのほうがレストランBよりも儲かっている」とは言い切れない
1. この折れ線グラフから、「この期間でレストランBは、月ごとの来客数が増加している」と言える
1. この折れ線グラフからは外部要因が分からないため、「7月には、食堂Aの来客数よりもレストランBの来客数の方が多くなる」とは言い切れない<br />ただし、食堂A、レストランBの来客数の変動が同じように続いた場合は、7月にはレストランBの来客数が食堂Aの来客数を超える確率が高いとはいい切れる
