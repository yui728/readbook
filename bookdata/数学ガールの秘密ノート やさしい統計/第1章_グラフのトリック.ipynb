## 1.2 表を作る

| 年 | 0  | 1  | 2  | 3  | 4  | 5  |
| -- | -- | -- | -- | -- | -- | -- |
| 社員数（人） | 100 | 117 | 126 | 133 | 135 | 136 |

<p style="font-weight: bold; text-align: center;">ユーリ株式会社の社員数</p>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('ggplot')

fig, ax = plt.subplots()
ax.set_title('社員数の折れ線グラフ')
ax.set_ylabel('社員数(人)')
ax.set_xlabel('年')
ax.set_ylim(0, 150)
ax.set_yticks([0, 50, 100])


# 年数
x = [0, 1, 2, 3, 4, 5]
# 社員数
y = [100, 117, 126, 133, 135, 136]

# 折れ線グラフを描画
plt.plot(x, y, marker = 'o')

plt.show()
```

## 1.3 大きく見せたい

- 大きく見せたい時は、グラフの下を切り取る

```python
import numpy as np

fig = plt.figure()
ax1 = plt.axes([0, 0.12, 1, 0.7])
ax2 = plt.axes([0, 0, 1, 0.1], sharex = ax1)
ax1.set_title('下を切り取った折れ線グラフ')
ax1.set_ylabel('社員数(人)')
ax2.set_xlabel('年')
ax2.set_ylim(0, 1)
ax2.set_yticks([0])
ax2.spines['top'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.set_ylim(90, 150)
ax1.set_yticks([100, 130])

ax1.xaxis.tick_top()
ax2.xaxis.tick_bottom()
ax1.tick_params(labeltop = 'false')
ax2.set_aspect('auto', adjustable = 'datalim')

d = .015
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs) 

d = .015
kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (0.8 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (0.8 - d, 1 + d), **kwargs)

# 年数
x = [0, 1, 2, 3, 4, 5]
# 社員数
y = [100, 117, 126, 133, 135, 136]

# 折れ線グラフを描画
ax1.plot(x, y, marker = 'o')

plt.subplots_adjust(hspace = 0.01)

plt.show()
```

- 省略するのではなく、下限を調整することでも大きく見せられる

```python
fig, ax = plt.subplots()
ax.set_title('省略を表す波線がない場合')
ax.set_ylabel('社員数(人)')
ax.set_xlabel('年')
ax.set_ylim(70, 150)
ax.set_yticks([70, 100, 130])


# 年数
x = [0, 1, 2, 3, 4, 5]
# 社員数
y = [100, 117, 126, 133, 135, 136]

# 折れ線グラフを描画
plt.plot(x, y, marker = 'o')

plt.show()
```

## 1.4 もっと大きく見せたい

- 切り取って引き伸ばすことで、更に大きく見せることができる

```python
fig = plt.figure(figsize = (3, 5))
ax1 = plt.axes([0, 0.105, 1, 0.7])
ax2 = plt.axes([0, 0, 1, 0.08], sharex = ax1)
ax1.set_title('切り取ってさらに引き伸ばした折れ線グラフ')
ax1.set_ylabel('社員数(人)')
ax2.set_xlabel('年')
ax2.set_ylim(0, 1)
ax2.set_yticks([0])
ax2.spines['top'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.set_ylim(95, 140)
ax1.set_yticks([100, 110, 120, 130])

ax1.xaxis.tick_top()
ax2.xaxis.tick_bottom()
ax1.tick_params(labeltop = 'false')
ax2.set_aspect('auto', adjustable = 'datalim')

d = .015
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs) 

d = .015
kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (0.8 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (0.8 - d, 1 + d), **kwargs)

# 年数
x = [0, 1, 2, 3, 4, 5]
# 社員数
y = [100, 117, 126, 133, 135, 136]

# 折れ線グラフを描画
ax1.plot(x, y, marker = 'o')

plt.subplots_adjust(hspace = 0.01)

plt.show()
```

- グラフには作る人の意図が入り込む
  - 大きく見せたい
  - $\therefore$グラフを読む側が注意深く読み取る必要がある
    - 切り取らなかったらどんなグラフだろうか
    - 拡大しなかったらどんなグラフだろうか

## 1.6 棒グラフ

## 1.7 横軸を変える

*横軸の目盛りをいじれば、社員が増えている印象にできる*

```python
fig = plt.figure(figsize = (3, 5))
ax1 = plt.axes([0, 0.105, 1, 0.7])
ax2 = plt.axes([0, 0, 1, 0.08], sharex = ax1)
ax1.set_title('社長「社員はすごく増えてるぞ」')
ax1.set_ylabel('社員数(人)')
ax2.set_xlabel('年')
ax2.set_ylim(0, 1)
ax2.set_yticks([0])
ax2.spines['top'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.set_ylim(95, 140)
ax1.set_yticks([100, 110, 120, 130])

ax1.xaxis.tick_top()
ax2.xaxis.tick_bottom()
ax1.tick_params(labeltop = 'false')
ax2.set_aspect('auto', adjustable = 'datalim')

d = .015
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs) 

d = .015
kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (0.8 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (0.8 - d, 1 + d), **kwargs)

# 年数
x = [0, 1, 2, 3, 4, 5]
# 社員数
y = [100, 117, 126, 133, 135, 136]

# 折れ線グラフを描画
ax1.plot(x, y, marker = 'o')

plt.subplots_adjust(hspace = 0.01)

plt.show()
```

*横軸の目盛りをいじっても、データの印象が変わる*

```python
fig, ax = plt.subplots()

# 年数
x = [d for d in range(6)]
# 社員数
y = [100, 117, 126, 133, 135, 136]

ax.set_title("専務「社員はほとんど増えてません」")
ax.set_xlabel("年")
ax.set_ylabel("増加人数（人）")
ax.set_xticks(x)
ax.set_ylim(0, 150)
ax.set_yticks([d * 50 for d in range(3)])

ax.plot(x[3:], y[3:], marker = 'o')

plt.show()
```

**グラフをどう描く**かだけではなく、**データのどこを選んでグラフを描くか**も大事
